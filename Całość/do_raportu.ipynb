```python
from importlib import reload

import numpy as np
from numpy.linalg import svd, norm
import pandas as pd
from time import time
from przypadek1 import *
from przypadek2 import *
from przypadek3 import *
from podziałB import *

import bidiagonalizacja
reload(bidiagonalizacja)
from bidiagonalizacja import *

import pelny_algorytm
reload(pelny_algorytm)
from pelny_algorytm import *
```

```python
M = [7, 10, 20, 50, 100, 200, 500, 1000, 2500]
N = [3, 7, 10, 20, 50, 100, 200, 300, 500]


timeDF = {}
SVaccuracy = {}
for m in M:
    for n in N:
        if m < n:
            continue
        print(f"{(m, n)=}")
        timeDF[(m, n)] = {"linalg": None, "dac": None}
        A = np.random.rand(m, n)
        start = time()
        trueSVD = svd(A)
        timeDF[(m, n)]["linalg"] = time()-start
        
        start = time()
        dacSVD = DACSVD(A)
        timeDF[(m, n)]["dac"] = time()-start
        
        if SVaccuracy.get(m) is None:
            SVaccuracy[m] = {n: 100*norm(dacSVD[1] - trueSVD[1])/norm(trueSVD[1])}
        else:
            SVaccuracy[m][n] = 100*norm(dacSVD[1] - trueSVD[1])/norm(trueSVD[1])
```

```python
timeDFpd = pd.DataFrame.from_dict(timeDF, orient="index")
print(
    timeDFpd.style.set_table_styles([
    {'selector': 'toprule', 'props': ':hline;'},
    {'selector': 'midrule', 'props': ':hline;'},
    {'selector': 'bottomrule', 'props': ':hline;'},
    ], overwrite=False).to_latex(clines="all;data")
)
```
